## Verification

```
import numpy as np
threshold = 1.5

!python test.py --cfg_path="./configs/arc_res50.yaml" --img_path="./data/steve.jpg"
encoded1 = np.load('./output_embeds.npy')

!python test.py --cfg_path="./configs/arc_res50.yaml" --img_path="./data/BruceLee1.jpg"
encoded2 = np.load('./output_embeds.npy')

diff = np.subtract(encoded1, encoded2)
dist = np.sum(np.square(diff), 1)
print(f"Steve Job vs Bruce Lee: {dist < threshold}")
```
